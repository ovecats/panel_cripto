```python
historical = requests.get('https://ftx.com/api/markets/BTC/USD/candles?resolution=3600&start_time=1609462800').json()
historical = pd.DataFrame(historical['result'])
historical.drop(['startTime'], axis = 1, inplace=True)
historical.head()
```
